**Арфметические операции**

Основной особенностью `ndarray` является то, как они работают с арифметическими операциями. Если взять два `ndarray` одинакового размера и применить к ним какую-нибудь арифметическую операцию, то получим новый массив, элементы которого получаются применением этой операции к соответствующим парам элементов.

Пример:
```python
a = np.array([20,30,40,50])
b = np.array([0, 1, 2, 3])
a - b
```
вывод:
```python
array([20, 29, 38, 47]).
```
_Проведите арифметические операции (не забудьте про +=, *=..) с произвольными массивами одинаковой формы._

````python
b1 = np.array([[1,2,3,91],[2,6,5,4],[1,45,21,54]])
b2 = np.array([[0,1,3,4],[1,1,3,2],[0,40,10,45]])

b1 += b1
print(b1)
print("\n")

b1 -= b2
print(b1)
print("\n")

b1 *= b2
print(b1)
print("\n")

print(b1 - b2)
print("\n")
print(b1 + b2)
print("\n")
print(b1 * b2)
````

Такая реализация математических операций позволяет записывать большинство вычислений над массивами без использования циклов, которые требуются в других языках программирования. Стиль написания программы, в котором используются подобные операции над целыми массивами называется **векторизованным кодом**. Векторизованный код гораздо компактнее обычного кода с циклами, и позволяет избежать ошибок с неправильной работой с индексами.

Как Вы могли заметить, операция умножения работает также поэлементно. Во многих случаях это удобно, но не следует путать такое умножение с умножением матриц в линейной алгебре, это две разные операции и есть случаи, в которых нужна именно вторая.

Чтобы перемножить 2 матрицы по правилам линейной алгебры, следует применить к ним функцию `np.dot()`:
```python
c = np.dot(a, b)  
```
Для одномерных массивов функция `np.dot()` вычисляет их скалярное произведение.

_Примените поэлементное и матричное умножение к двум матрицам._

````python
b1 = np.array([2,3,4,5,1])
b2 = np.array([4,2,4,2,1])
b3 = np.dot(b1,b2)
print(b3)
b4 = b1 * b2
print(b4)
````

При этом будет сформирован новый массив, каждый элемент которого будет равен результату выполнения операции между соответствующим элементом исходного массива и числом. **Возможно проблема в том, что мы не можем привести массив чисел с плавающей точкой к массиву с целыми числами**

Обычно вычисления над элементами массива не ограничиваются только четырьмя арифметическими операциями. В библиотеке Numpy реализованы все стандартные математические функции, которые можно применить сразу ко всем элементам массива, например в результате выполнения кода
```python
a = np.arange(1, 5)
np.sqrt(a)
```
будут вычислены квадратные корни всех элементов массива `a`:
```python
array([ 1.,  1.41421356,  1.73205081,  2.])
```

Такие функции в документации Numpy называют универсальными (анг. universal function, сокращённо ufunc).

_Выберите пару (или больше) функций из https://docs.scipy.org/doc/numpy/reference/routines.linalg.html или https://docs.scipy.org/doc/numpy/reference/routines.math.html и проиллюстрируйте их работу в окне ниже._

````python
b1 = np.array([2,10,2,6])
b2 = np.array([-2,10,4,6])

b3 = np.inner(b1,b2)
print(b3)

c1 = np.array([[2,2],[4,5]])
c2 = np.array([[4,5],[2,1]])
b4 = np.linalg.inv(c1)
print(b4)

b5 = np.linalg.det(c1)
b5
````

**inner= Умножает первый элемент из первого массива с первым элементом из второго массива и суммирует с произведением вторых элементов двух массивов и так далее до конца массива**

**inv = находит обратную матрицу для заданного массива**

**det = находит детерминант  для заданного массива**

Ещё один вид функций, который определён в Numpy, это функции вычисляющие агрегированные показатели по всем элементам массива. Они реализуют такие операции как вычисление суммы всех элементов (функция `np.sum`), нахождения минимального (`np.min`) и максимального (`np.max`) элементов, и многие другие.

Такие функции обычно можно вызвать двумя способами, первый как методы класса `ndarray`, например:
```python
a.sum()
a.min()
a.max()
```

Второй способ как обычную функцию библиотеки Numpy:
```python
np.sum(a)
np.min(a)
np.max(a)
```

Второй вариант позволяет передать функции не только `ndarray`, но и обычные списки Python или любые другие "подходящие" массивы. В этом случае входной массив будет преобразован в `ndarray` перед выполнением операции.
Некоторые из функций могут быть доступны только в одном из вариантов.

**Индексация и срезы по массивам**

Для доступа к элементам массивов `ndarray` используются квадратные скобки. К примеру, `a[2]` -- второй элемент массива `a`.
Таким образом, доступ к элементам одномерных массивов такой же, как и к классическим спискам (`list`) Python.

Также удобно использовать срезы (в квадратных скобках можно указать `начало : конец : шаг`):
```python
a = np.arange(10)**2
print(a[2])
print(a[2:5])
a[:6:2] = -5  #эквивалентно a[0:6:2] = -5 -- с начала до 6 позиции с шагом 2 (каждому второму элементу) устанавливается значение -5
```
Чтобы обратиться к последнему элементу, используйте индекс "-1".

Для того, чтобы передвигаться по массиву в обратном порядке (с конца в начало), в качестве шага также можно использовать -1: `a[ : :-1]`.

````python
a = np.arange(10)**2
print(a)
a[::-1]

aa = a[:7:2]
print(aa)
print(a[7])
````

Доступ к элементам многомерных массивов осуществляется через выбор индекса по каждой оси. Индексация в этом случае производится через запятую внутри двойных квадратных скобок: 
- `a[0, 2]` -- нулевая строка и второй столбец массива `a`, 
- `a[1:3, 2]` -- элементы из 1 и 2 строки второго столбца.
- `a[:, 2]` -- элементы из всех строк 2го стобца.

Если количество индексов меньше числа осей, то недостающие индексы автоматически заполняются двоеточием:
- `a[1]` -- все числа из 1 строки матрицы `a`.

Для "недостающих" индексов также наравне с двоеточием можно использовать троеточие:
```python
a = np.array( [[[ 1, 2, 3], [ 11, 12, 13]], 
                [[101,102,103], [111,112,113]]]) #трёхмерный массив (2 двумерных массива)
а[1,...]   # аналогично a[1,:,:] или a[1]
а[...,2]   # аналогично a[:,:,2]
```

Обратите внимание, что **срезы не создают новый массив при работе с numpy-массивами**. Они возвращают специальный объект, который работает с теми же данными, которые находятся в исходном массиве. Например, следующий код изменяет данные массива `a` через массив `b`:
```Python
a = np.array([1, 2, 3, 4])
b = a[::2]
b[1] = 0
a
```
В результате будет выведено:
```Python
array([1, 2, 0, 4])
```
**Такое поведение отличается от поведения обычных list-списков Python - в них срезы создают копию данных**. 

Данная особенность срезов `ndarray` может привести к неожиданным (и неприятным!) последствиям, такие ошибки в коде сложно отловить. Чтобы обезопасить себя, в таких случаях рекомендуется делать независимую копию данных -- для этого можно использовать функцию `np.array` или метод `np.copy()`. **Конкатенация массивов**

При обработке данных часто требуется собирать большие массивы из фрагментов меньшего размера. Функцией, выполняющей соединение (конкатенацию) массивов является `np.concatenate`:
```Python
np.concatenate((a1, a2, ...), axis=0)
```

Параметр `axis` определяет ось, по которой будет выполнена конкатенация. По умолчанию `axis=0` что соответствует конкатенации строк:
```Python
a = np.array([[1,2], [3,4]])  #a.shape: (2, 2)
b = np.array([[5, 6]])        #b.shape: (1, 2)
np.concatenate((a, b), axis=0)
```
Выведет:
```Python
array([[1, 2],
       [3, 4],
       [5, 6]])
```
Для того, чтобы конкатенация сработала, все соединяемые матрицы должны иметь совпадающие размеры по совмещаемым осям (для оси, используемой для объединения, это не важно). Поэтому, обратите внимание, в примере вектор `b` является двумерной матрицей(!), так же как и `a`.

Чтобы выполнить *конкатенацию по столбцам* следует задать параметр `axis=1`:
```Python
np.concatenate((a, b.T), axis=1)
```
Получим:
```Python
array([[1, 2, 5],
       [3, 4, 6]])
```

Обратите внимание, что при конкатенации по столбцам в этом примере необходимо транспонировать матрицу `b`, так как иначе матрицы будут не совместимы друг с другом по форме. 

Функция `np.stack` соединяет матрицы вдоль нового измерения. Например, с её помощью можно, склеить двумерную матрицу из одномерных векторов - строк:

```Python
u = np.array([1, 2])
v = np.array([3, 4])
w = np.array([5, 6])
np.stack((u, v, w))
```
Получим:
```Python
array([[1, 2],
       [3, 4],
       [5, 6]])
```

Функции `np.vstack` и `np.hstack` соединяют массивы вертикально и горизонтально соответственно. Эти функции считаются устаревшими, однако последняя из них удобна тем, что позволяют присоединить вектор к матрице не увеличивая предварительно размерность вектора до двух. Это может быть полезно в ситуции, когда строки матрицы вычисляются последовательно и нужно дописывать их к ранее накопленным данным.

```Python
M = np.array([[1, 2], [3, 4]])
w = np.array([5, 6])
np.vstack((M, w))
```
Получим:
```Python
array([[1, 2],
       [3, 4],
       [5, 6]])
```
И наконец, функция `np.dstack` использовалась для соединения двумерных массивов (например, изображений) по третьему измерению. Вместо этой функции теперь также рекомендуется использовать `np.stack`.
